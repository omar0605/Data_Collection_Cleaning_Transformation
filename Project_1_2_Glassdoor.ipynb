```python
import numpy as np
import pandas as pd
```

```python
df_glsdr = pd.read_csv(r'D:\Python Code\Python-2\PortFolio\Project_1_2_Glassdoor\Uncleaned_DS_jobs.csv')
df_glsdr.head()
```

```python
df_glsdr = df_glsdr.drop_duplicates()
df_glsdr.head()
```

```python
df_glsdr[["Salary Estimate min","Salary Estimate max"]]=df_glsdr["Salary Estimate"].str.split('-',2,expand=True)
df_glsdr[["Salary Estimate min","Salary Estimate max"]].head()
```

```python
df_glsdr["Salary Estimate min"]= df_glsdr["Salary Estimate min"].str.lstrip('$')
df_glsdr["Salary Estimate min"]= df_glsdr["Salary Estimate min"].str.rstrip('K')
df_glsdr["Salary Estimate min"]= df_glsdr["Salary Estimate min"].astype(int)
df_glsdr[["Salary Estimate min","Salary Estimate max"]].head()
```

```python
df_glsdr[["Salary Estimate min","Salary Estimate max"]].head()
```

```python
df_glsdr["Salary Estimate max"] = df_glsdr["Salary Estimate max"].str.strip('(Glassdoor est.)K$" "')
df_glsdr["Salary Estimate max"]= df_glsdr["Salary Estimate max"].str.strip("...")
df_glsdr["Salary Estimate max"] = df_glsdr["Salary Estimate max"].str.strip('K(Employ')
df_glsdr["Salary Estimate max"] = df_glsdr["Salary Estimate max"].astype(int)
```

```python
df_glsdr[["Salary Estimate min","Salary Estimate max"]] = df_glsdr[["Salary Estimate min","Salary Estimate max"]]*1000
df_glsdr[["Salary Estimate min","Salary Estimate max"]].head()
```

```python
df_glsdr[["Company Name","Col no need"]]= df_glsdr["Company Name"].str.split("\n", expand=True)
```

```python
df_glsdr["Company Name"] = df_glsdr["Company Name"].str.strip('1234567890.-@$#')
df_glsdr.head()
```

```python
df_glsdr[["City","State","Col no eed2"]] = df_glsdr["Location"].str.split(',', expand=True)
```

```python
df_glsdr[["City","State"]]
df_glsdr["State"] = df_glsdr["State"].str.strip('...')
df_glsdr["City"] = df_glsdr["City"].str.rstrip('...')
df_glsdr[["City","State"]]
```

```python
df_glsdr["Job Description"].str.strip('...')
df_gls = df_glsdr["Job Description"].str.split('\n',expand=True)
df_gls = df_gls.replace('Duties','')
df_gls
```

```python
df_gls["Job Position"]=df_gls[1]+df_gls[2]+df_gls[3]+df_gls[4]
df_gls["Duties/Responsibilities"]=df_gls[3]+df_gls[4]+df_gls[5]
df_gls[["Job Position","Duties/Responsibilities"]]
#df_gls["Duties"]=df_gls["Duties/Responsibilities"].str.split(':')
df_gls["Duties/Responsibilities"].str.replace('Duties','')
```
